```python
import numpy as np
import matplotlib.pyplot as plt

img1 = plt.imread('image/bzebra.jpg')
img2 = plt.imread('image/nature.jpg')

#get the gray Scale image
img1 = np.mean(img1, axis=-1)
img2 = np.mean(img2, axis=-1)

height = max(img1.shape[0],img2.shape[0])
width = max(img1.shape[1],img2.shape[1])
print(img1.shape)
print(img2.shape)
print('Result image size will : ',height, width)

reg_img = np.zeros((height,width), dtype=np.uint8)
for i in range(height):
    for j in range(width):
        px = 0
        if i<img1.shape[0] and j<img1.shape[1]:
            px=int(img1[i][j]*0.35)
        if i<img2.shape[0] and j<img2.shape[1]:
            px+=img2[i][j]
        reg_img[i][j] = px%255

print(reg_img.shape)

# fig, axis = plt.subplots(1,3)

# axis[0].imshow(img1, cmap='gray')

# axis[1].imshow(img2, cmap='gray')

# axis[2].imshow(reg_img, cmap='gray')

# # plt.imshow(reg_img, cmap='gray')

plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
plt.imshow(img1, cmap='gray')
plt.title('Image A')
plt.subplot(1,2,2)
plt.title('Image B')
plt.imshow(img2, cmap='gray')
plt.show()
            
plt.imshow(reg_img, cmap='gray')
plt.title('Addition Image')
```
