```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load an example image
image_path = 'image/natures.png'  # Replace with the path to your image
original_image = plt.imread(image_path)
gray_image = np.mean(original_image, axis=-1)
plt.imshow(gray_image, cmap='gray')
plt.title('Input Images')
```

```python
# Apply contrast stretching using loops
lsb_img = np.zeros_like(gray_image)
for i in range(lsb_img.shape[0]):
    for j in range(lsb_img.shape[1]):
        lsb_img[i, j] = np.clip(((int(gray_image[i,j]*255) & (1 << 0)) != 0)*255,0,255)
plt.imshow(lsb_img, cmap='gray')
plt.title('Lsb Image')
```

```python
number_of_bit = 8
# Create a figure and subplots
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(number_of_bit, 4))

# Flatten the axes array for easier iteration
axes = axes.flatten()


# Plot each grayscale image on a subplot
for ii, ax in enumerate(axes):
    # Apply contrast stretching using loops
    lsb_img = np.zeros_like(gray_image)
    for i in range(lsb_img.shape[0]):
        for j in range(lsb_img.shape[1]):
            lsb_img[i, j] = np.clip(((int(gray_image[i,j]*255) & (1 << ii)) != 0)*255,0,255)
            # Display the images
    ax.imshow(lsb_img, cmap='gray')  # Display each grayscale image
    ax.set_title(f'Image {i + 1}')
    ax.axis('off')  # Turn off axis labels for better visualization

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
