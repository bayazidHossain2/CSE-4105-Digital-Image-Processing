# 04_add_the_boundary_assign_it_to_the_center

```python
import numpy as np
import matplotlib.pyplot as plt

img = plt.imread('image/bzebra.jpg')
#get the gray Scale image
img = np.mean(img, axis=-1)

height, width = img.shape
sum_center = np.zeros((height, width),dtype=np.uint8)
# sum_top_left = np.zeros((height, width),dtype=np.uint8)
# sum_center = np.zeros((height, width),dtype=np.uint8)
# sum_center = np.zeros((height, width),dtype=np.uint8)
space = 2
for i in range(1,height-1):
    for j in range(1,width-1):
        if(i%space==0)and(j%space==0):
            region = img[i-1:i+2,j-1:j+2]
            sum = np.sum(region)%256
            # print(region)
            # print(sum)
            sum_center[i][j] = sum
        else:
            sum_center[i][j] = img[i][j]
        # break
    # break
plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(sum_center, cmap='gray')
plt.show()
```
