# 03 Get Diagonal value of the image, assign them as zero, 

```python
import numpy as np
import matplotlib.pyplot as plt

img = plt.imread('image/bzebra.jpg')
#get the gray Scale image
img = np.mean(img, axis=-1)
print(img.shape)
height, width = img.shape
min_axis = min(height, width)
print(min_axis)
square_img = img[0:min_axis, 0:min_axis]
print(square_img.shape)
for i in range(height):
    for j in range(width):
        if(i == j):
            img[i][j] = 0
        elif(i+j == min_axis):
            img[i][j] = 0

plt.imshow(square_img, cmap='gray')
```
