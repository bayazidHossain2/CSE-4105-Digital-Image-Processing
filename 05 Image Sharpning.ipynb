```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load an example image
image_path = 'image/nature256.png'  # Replace with the path to your image
original_image = plt.imread(image_path)
gray_image = np.mean(original_image, axis=-1)
plt.imshow(gray_image, cmap='gray')
plt.title('Input Images')
```

```python

# Define the smoothing filter (e.g., Gaussian blur)
blur_kernel = np.array([[1, 2, 1],
                        [2, 4, 2],
                        [1, 2, 1]]) / 16  # Gaussian blur kernel

# Apply the smoothing filter using convolution
smoothed_img = np.zeros_like(gray_image)
for i in range(1, gray_image.shape[0] - 1):
    for j in range(1, gray_image.shape[1] - 1):
        smoothed_img[i, j] = np.sum(gray_image[i-1:i+2, j-1:j+2] * blur_kernel)

print('Max : ', np.max(smoothed_img), ' Min : ', np.min(smoothed_img))
plt.imshow(smoothed_img, cmap='gray')
plt.title('Smooth Image')
```

```python
# Define the sharpening kernel
sharpen_kernel = np.array([[0, -1, 0],
                        [-1, 5, -1],
                        [0, -1, 0]])

# Apply the sharpening kernel using convolution
sharpened_img = np.zeros_like(smoothed_img)
for i in range(1, smoothed_img.shape[0] - 1):
    for j in range(1, smoothed_img.shape[1] - 1):
        sharpened_img[i, j] = np.sum(smoothed_img[i-1:i+2, j-1:j+2] * sharpen_kernel)

# Clip values to ensure they are within valid range [0, 255]
# sharpened_img = np.clip(sharpened_img, 0, 255).astype(np.uint8)

print('Max : ', np.max(sharpened_img), ' Min : ', np.min(sharpened_img))
plt.imshow(sharpened_img, cmap='gray')
plt.title('Sharped Image')
```
