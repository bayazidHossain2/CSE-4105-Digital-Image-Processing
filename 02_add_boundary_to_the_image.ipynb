# 02 Add bounday to the input image

```python
import numpy as np
import matplotlib.pyplot as plt

img = plt.imread('image/bzebra.jpg')
# img = np.mean(img, axis=-1)
padding_size = 10
color = 0
print(img.shape)

if(len(img.shape) == 3):
    print('Color image')
    height, width, channel = img.shape
    padded_img = np.full((height + 2 * padding_size, width + 2 * padding_size, channel), color, dtype=img.dtype)
    print(padded_img.shape)
    padded_img[padding_size:-padding_size, padding_size:-padding_size] = img
    # print(padded_img)
elif(len(img.shape) == 2):
    print('Gray Scale image')
    height, width = img.shape
    padded_img = np.full((height + 2 * padding_size, width + 2 * padding_size), color, dtype=np.uint8)
    print(padded_img.shape)
    padded_img[padding_size:-padding_size, padding_size:-padding_size] = img
    # print(padded_img)

plt.figure(figsize=(8,3))
plt.subplot(1, 2, 1)
plt.title('Original Image')
if(len(img.shape)==2):
    plt.imshow(img, cmap='gray')
else:
    plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Padded image')
if(len(img.shape)==2):
    plt.imshow(padded_img, cmap='gray')
else:
    plt.imshow(padded_img)
# plt.imshow(padded_img)

plt.show()
```
