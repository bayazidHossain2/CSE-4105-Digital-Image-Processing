# 13 Contrast stretching

```python
import numpy as np
import matplotlib.pyplot as plt

img = plt.imread('image/nature.jpg')
img = np.mean(img, axis = -1)

r1=100
r2=200
s1=120
s2=240
alpha = (s1/r1)
bita = (s2-s1)/(r2-r1)
gama = (255-s2)/(255-r2)

height, width = img.shape
stratched_img = np.zeros((height, width), dtype=np.uint8)
for i in range(height):
    for j in range(width):
        if(img[i][j]<r1):
            stratched_img[i][j] = int(img[i][j]*alpha) % 256
        elif(img[i][j]<r2):
            stratched_img[i][j] = int(img[i][j]*bita) % 256
        else:
            stratched_img[i][j] = int(img[i][j]*gama) % 256


plt.figure(figsize=(8,3))
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original image')
plt.subplot(1,2,2)
plt.imshow(stratched_img, cmap='gray')
plt.title('Contrast Stretched Image')


print(img.shape)
```

```python
nk = np.zeros(256)
nks = np.zeros(256)
for i in range(height):
    for j in range(width):
        nk[int(img[i][j])] += 1
        nks[int(stratched_img[i][j])] += 1


plt.bar(np.arange(256), nk)
plt.bar(np.arange(256), nks)
```
